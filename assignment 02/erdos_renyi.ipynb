```python
import networkx as nx
from itertools import combinations
import random 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Input parameters.
n = 1000
p = 0.5
```

```python
# Generate iterator over all node combinations.
possible_edges = combinations(range(n), 2)
edge_list = [edge for edge in possible_edges if random.random()<p]
g = nx.Graph()
g.add_nodes_from(range(n))
g.add_edges_from(edge_list)
```

```python
# Plot degree histogram.
degrees = nx.degree_histogram(g)
plt.bar(list(range(len(degrees))), degrees, align='center')
```

```python
# Analyze numbers of isolated nodes.
nodes = 1000
p_window = list(np.linspace(0.0000000000005,0.003, num=500))
values = []
for prob in p_window:
    g = nx.erdos_renyi_graph(nodes, prob)
    num_isolates = nx.number_of_isolates(g)
    values.append(num_isolates)
plt.plot(p_window, values)
```
