```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
def plot_degrees(graph):
    # Plot degree histogram.
    degrees = nx.degree_histogram(graph)
    plt.bar(list(range(len(degrees))), degrees, align='center')
```

```python
NUM_NODES = 3000
NUM_EDGES_TO_ADD = 5

# Create random barabasi-albert graph.
bar_alb_graph = nx.barabasi_albert_graph(NUM_NODES, NUM_EDGES_TO_ADD)
print(f"Barabasi-Albert graph has: {bar_alb_graph.number_of_nodes()} nodes, {bar_alb_graph.number_of_edges()} edges.")
#nx.draw_networkx(bar_alb_graph)
plot_degrees(bar_alb_graph)
```

```python
# Generate ER graph with same size.
NUM_NODES = bar_alb_graph.number_of_nodes()
NUM_EDGES = bar_alb_graph.number_of_edges()
erd_renyi_graph = nx.gnm_random_graph(NUM_NODES, NUM_EDGES)
print(f"Erdos-Renyi graph has: {erd_renyi_graph.number_of_nodes()} nodes, {erd_renyi_graph.number_of_edges()} edges.")
plot_degrees(erd_renyi_graph)
```
